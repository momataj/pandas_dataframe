```python
import pandas as pd
```

```python
train = pd.read_csv('http://bit.ly/kaggletrain')
```

```python
train.head()
```

```python
#Map method
```

```python
#dummy variable and apply map. Map allow you to exiting data in the series
```

```python
train['Sex_num']=train.Sex.map({'female':0,'male':1})
```

```python
train.loc[0:4,['Sex','Sex_num']] #cloumns name with loc
```

```python
#apply a function in each element of series  Series Method
```

```python
train['Name_length']=train.Name.apply(len)  # apply for apply function
```

```python
train.loc[0:4,['Name','Name_length']]
```

```python
import numpy as np
```

```python
#ceiling function
```

```python
train['Fare_ceil']=train.Fare.apply(np.ceil)
```

```python
train.loc[0:4,['Fare','Fare_ceil']]  # ceil function rounding up
```

```python
#last name , part before the comma
```

```python
train.Name.str.split(',').head()   # split the name with comma
```

```python
def get_element(my_list,postion):
    return my_list[postion]    # return the postion of name from the list
```

```python
train.Name.str.split(',').apply(get_element,postion=0).head()
```

```python
train.Name.str.split(',').apply(lambda x:x[0]).head()  #using the Series method apply lamda function
```

```python
drinks=pd.read_csv('http://bit.ly/drinksbycountry')
```

```python
drinks.head()
```

```python
drinks.loc[:,'beer_servings':'wine_servings'].apply(max,axis=0)  # max value for each columns
```

```python
drinks.loc[:,'beer_servings':'wine_servings'].apply(max,axis=1) #max value for eacch raws
```

```python
drinks.loc[:,'beer_servings':'wine_servings'].apply(np.argmax,axis=0)  #ewhich column is the maxi
```

```python
drinks.loc[:,'beer_servings':'wine_servings'].apply(np.argmax,axis=1)  #which columns is the max
```

```python
drinks.loc[:,'beer_servings':'wine_servings'].apply(np.argmax,axis=0) #which raw is the max
```

```python
drinks.loc[:,'beer_servings':'wine_servings'].applymap(float)    # apply every element of the data frame 
```

```python
drinks.loc[:,'beer_servings':'wine_servings']=drinks.loc[:,'beer_servings':'wine_servings'].applymap(float)
drinks.head()  # apply map in all raws in the data frame , convert int to float
```
